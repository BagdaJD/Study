*1)Вычисление машинного эпсилона(Алгоритм от Багды)*

```python
#В конспектах не нашел
n = 5 #нужная нам точность
eps = 1

for _ in range(n):#вычисляем эпсилон с заданной точностью (n)
    eps /= 10
print(eps)
#Но это не точно
```

*10)Вычисление 2^n*

```python
#Алгоритм от Багды(в конспектах не нашел)

n = int(input()) #степень

a = 2
for _ in range(n - 1):
    a *= 2

print(a)
```

*11) Вычисление n!!*




```python
#Алгоритм от Багды, вроде работает, но в конспектах я его не нашел
n = int(input())

fact = 1
for i in range(n, 0, -2):
    fact *= i

print(fact)
```
