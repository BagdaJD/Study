```python
#  решаем пункт 1

from sympy import Symbol, Sum

n = int(input('n = '))

z = -1
s = z / 3
for i in range(2, n+1):
    z = -z
    s = s + z / (2 * i + 1)
    
print('summa = %7.5f' %s)

#  проверка
i = Symbol('i')
ss = Sum((-1) ** i / (2 * i + 1), (i, 1, n))
print('sympy = %7.5f' %ss)
```

# Примеры вычисления $x^n$ и $n!$ (рекуррентно)

```python
x = float(input('x = '))
n = int(input('n = '))  # натуральное, т.е. n = 1, 2, 3,...

# y = x^n
y = x    #  при n = 1
for _ in range(2, n+1):
    y = y * x
    
print(y)
```

```python
n = int(input('n = '))  # натуральное, т.е. n = 1, 2, 3,...

# y = n!
y = 1    #  при n = 1
for i in range(2, n+1):
    y = y * i
    
print(y)
```

```python
#  решаем пункт 1

from math import sin as msin
from sympy import Symbol, Product, sin, factorial

x = float(input('x = '))
n = int(input('n = '))

# начальные значения при i = 1
fact = 1    # i! = 1
p = 1 + msin(x) / fact
for i in range(2, n+1):
    fact = fact * i
    p = p * (1 + msin(i * x) / fact)
    
print('prod = %7.5f' %p)


#  проверка
i = Symbol('i')
ps = Product(1 + sin(i * x) / factorial(i), (i, 1, n))
print('symp = %7.5f' %ps)
```

Считаем $\sum\limits_{i=1}^n y_i$, где $y_i = \sin^i x$. 
При этом каждое слагаемое вычисляем рекуррентно: $y_i = C \cdot y_{i-1} = y_i = \sin x \cdot y_{i-1}$

```python
#  решаем пункт 1

from math import sin as msin
from sympy import Symbol, Sum, sin

x = float(input('x = '))
n = int(input('n = '))

# начальные значения при i = 1
y = msin(x)
s = y
for _ in range(2, n+1):
    y = y * msin(x)
    s = s + y

print('summa = %7.5f' %s)

#  проверка
i = Symbol('i')
ss = Sum(sin(x) ** i, (i, 1, n))
print('sympy = %7.5f' %ss)
```
