## 19s.2. Поиск максимального/минимального из двух выражений

**Постановка задачи**: даны выражения x и y. Найти $\max(x; y)$, $\min(x; y)$.

**I способ** (if)

Этот способ легко преобразуется в алгоритм (программу) поиска максимума/минимума среди любого количества элементов.

 ***Поиск $max(x; y)$*** (if)
 
```
max_ = x
if y > max_: 
    max_ = y
```


***Поиск $min(x; y)$*** (if)

```
min_ = x
if y < min_: 
    min_ = y
```

**II способ** (if-else)

Способ удобен для инициализации максимума/минимума при поиске максимума/минимума среди любого количества элементов.
```
if x > y: 
    max_ = x
else: 
    max_ = y
```

**III способ** (без использования условного оператора)

$$
  \max(x; y) = \frac{x + y + |x − y|}{2},\quad
  \min(x; y) = \frac{x + y − |x − y|}{2}
$$

**Осторожно!** Возможно переполнение для целых $x$ и $y$! (не Python)

````python
x, y = 12, -2
mx = (x + y + abs(x - y)) / 2
mn = (x + y - abs(x - y)) / 2
print(f'mx = {mx},   mn = {mn}')
````

### Алгоритм одновременного поиска $\max(x;y)$ и $\min(x;y)$

**Конструкция if-else**
```
if x < y:
    min_ = x
    max_ = y
else:
    min_ = y
    max_ = x
```

**Конструкция if**
```
min_ = x
max_ = y
if max_ < min_:
    min_ = y
    max_ = x
```

## 19s.3.  Поиск максимального/минимального из трех выражений

**Постановка задачи**: даны выражения $x$, $y$, $z$. Найти
$\max(x; y; z)$; $\min(x; y; z)$.

**Алгоритм (на примере максимума).**

* 1) Найти $\max(x; y)$, запомнить в переменную: max_xy := $\max(x; y)$;

* 2) Найти максимум от найденного в 1) значения и $z$: max(max_xy; z).

**Пример поиска $\max(x; y; z)$** <a id='section_id1'></a>

````python
x = int(input('x >>> '))
y = int(input('y >>> '))
z = int(input('z >>> '))

# поиск max(x; y)
if x > y:
    max_ = x
else:
    max_ = y

# поиск max(max(x; y); z)
if z > max_:
    max_ = z

print('max(%d, %d, %d) = %d' % (x, y, z, max_))
````

## 19s.4. Поиск $\max()$. Использование встроенных функций языка Python

````python
# max(x; y)
x, y = 5, 10
max_ = max(x, y)
print('max(%d, %d) = %d' % (x, y, max_))


# min(x; y; z);  max(x; y; z)
x, y, z = 15, 1, 7
min_ = min(x, y, z)
max_ = max(x, y, z)
print('min(%d, %d, %d) = %d' % (x, y, z, min_))
print('max(%d, %d, %d) = %d' % (x, y, z, max_))


# max(<строка>)
s = 'cat and dog' 
max_ = max(s)
print('max(%s) = %s' % (s, max_))


# min(<список чисел>)
a = [67, 89, 71]
min_ = min(a)
print('min(%s) = %d' % (a, min_))


# min(<список строк>)
a = ['cat', 'io', 'Cat'] 
min_ = min(a)
print('min(%s) = %s' % (a, min_))
````

## 19s.5. Тернарная условная операция

Простейшую условную инструкцию

```
if X:
    A = Y
else:
    A = Z
```
можно сократить до одной строки, используя следующую конструкцию:

`A = Y if X else Z`

В результате работы инструкции, выполнится выражение Y, если значение X
истинно, иначе выполнится выражение Z.

**Поиск $\max(x; y)$ с использование условной операции**

`max_ = x if x > y else y`

**Примеры использование условной операции**

````python
x = int(input('int x >>> '))

# Проверка числа на чётность; условный оператор
if x % 2 == 0: 
    print('Even')   # чётное
else: 
    print('Odd')  # нечётное
    
# --------------------------------------

# Проверка числа на чётность; условная операция. Способ 1
print('Even') if x % 2 == 0 else print('Odd')

# --------------------------------------

# Проверка числа на чётность; условная операция. Способ 2
print('Even' if x % 2 == 0 else 'Odd')
````

````python
# Поиск max(x,y) и min(x,y); условная операция

x = int(input('x >>> '))
y = int(input('y >>> '))

max_, min_ = (x, y) if x > y else (y, x)

print('Min =', min_)
print('Max =', max_)
````

Тернарная условная операция может использоваться внутри других выражений. Например,

````python
x = int(input('x >>> '))

R = x + (100 if (x % 2 == 0) else -100)
print('R =', R)

#В этом случае условную операцию желательно заключать в скобки.
````
