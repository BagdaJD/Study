# Алфавит и словарь языка Python

**Алфавит языка программирования** (ЯП) — набор допустимых символов, которые используются для записи программы.

**Лексема** (token) — минимальная единица языка, имеющая самостоятельный смысл.

Виды лексем:
- Литералы (literals);
- Операции (operators);
- Знаки пунктуации (разделители, punctuators);
- Ключевые слова (keywords);
- Идентификаторы (identifiers).

Лексемы формируют **словарь языка**.

 ### Представление чисел в ЯП
 
 В вещественном числе целая часть от дробной отделяется точкой.

Вещественные числа имеют две формы записи:

------------------------


с фиксированной точкой|с плавающей точкой
----------------|-------
0.13 | 5E-8
3.1415 | 9.422e+08
0.5 | 1E10
7.0  |

Цифру 0 до/после точки лучше не опускать. Регистр буквы «e» не имеет значения

Числа с плавающей точкой имеют *мантиссу* (число с фиксированной точкой) и *масштабный множитель* (порядок).

Например, запись `9.422E08` означает, что число $9.422$ надо умножить на $10^8$.

Здесь `9.422` — мантисса; `08` — порядок.

*Порядок* — это целое число со знаком. Если порядок положительный, то знак «+» можно опускать.

Идентичные варианты записи числа $0{,}051$:
```
0.051
0.051E0
0.51E-1
5.1E-2
...
```

**Предпочтительная форма записи числа с плавающей точкой**: мантисса вещественного числа должна принадлежать интервалу $[1; 10)$.


Представление числа в форме с плавающей точкой соответствует компьютерной реализации экспоненциальной записи чисел:
$$
N = M \cdot 10^p;
$$
$M$ — мантисса; $p$ ($∈ Z$) — порядок.

**!!!** Экспоненциальная запись удобна при представлении очень больших и очень малых чисел, а также для унификации написания чисел.

### Выражения

**Выражения** строятся из чисел и переменных с помощью знаков арифметических операций,

круглых скобок и функций.

Знак операции определяет действие над операндами:

`<I операнд> <знак операции> <II операнд>`

Примеры: `a + 4`;  `2 / 4`

Знак «*» опускать или заменять точкой нельзя.

Не рекомендуется размещать два знака операций рядом

Переменные и числа — простейшие частные случаи выражений.

## Операции

`+ - * ** / // % @ << >> & | ^ ~ < > <= >= == !=`

При вычислении значений выражений действуют обычные правила старшинства операций (приоритет операций).

**Приоритет операций** — очерёдность выполнения операций в выражении при отсутствии явного указания на порядок их вычисления (с помощью круглых скобок).

Подряд идущие операции одного приоритета вычисляются в последовательности «слева направо» (**ассоциативность слева**).

Исключения: в языке Python операция возведения в степень (`**`) **ассоциативна справа**.


````python
# Чему равно значение выражения 2+2*3?
print(2+2*3)

# Чему равно значение выражения 2**2**3?
print(2**2**3)
````

**Унарные операции** требуют одного операнда (аргумента операции): `-b`;  `not B1`.

**Бинарные операции** требуют двух операндов:

`a + b`

`B1 or B2`

* арифметические операции (самый высокий приоритет): `**`; `~`; унарные `+ -` `*`; ...
* операции сравнения: `<= < >` ...
* логические операции (самый низкий приоритет): `not`; `and`; `or`

Строки исходного файла: 

+ **физические** (в конце признак конца строки);

+ **логические** (для интерпретатора Python программа состоит из логических строк). Содержат одну или более физических строк, соединяющихся правилами языка.

````python
# одна логическая строка
print('Начало строки продолжение строки')
````

````python
# Плохая попытка разбить логическую строку на две физические
print('Начало строки
      продолжение строки')
````

````python
# Хорошая попытка разбить логическую строку на две физические
print('Начало строки \
продолжение строки')
````

````python
'''
Задание L4.1. С клавиатуры вводится целое трёхзначное число x.
Найти сумму цифру числа x, если число кратно трём и 
оканчивается цифрой 7. Если условие не выполняется, то
вывести текст No.
'''

xx = int(input('x >>> '))

x = abs(xx)

if x % 3 == 0 and x % 10 == 7:
    x0 = x % 10         # цифра разряда единиц
    x1 = x // 10 % 10   # цифра разряда десятков
    x2 = x // 100       # цифра разряда сотен

    sum_ = x0 + x1 + x2 
    print(f'Для числа {xx} ответ {sum_}')
else:
    print('No')

 
'''
Многострочные комментарии с тестовыми примерами
125 - - No
127 - + No
123 + - No
147 + + 12
'''  

# Пример цикла и списков
test  = [125, 127, 123, 147] 
print('-' * 24)
for t in test:
    print(f'{t} | {str(t % 3 == 0):7s} | {t % 10 == 7:d} |')

    print('%d | %-7s | %d |' % (t, t % 3 == 0, t % 10 == 7))

print('fff %d' % 12)
````
