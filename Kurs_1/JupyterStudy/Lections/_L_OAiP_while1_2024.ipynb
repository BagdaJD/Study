## 3o.1 Циклы

**Циклический алгоритм (цикл)** — алгоритм, у которого выполнение инструкций осуществляется многократно с одними и теми же или модифицированными данными.

* * *

**Заголовок цикла и тело цикла**

**Заголовок цикла** задает правила, определяющие число повторений тела цикла.

**Тело цикла** — последовательность инструкций, предназначенная для многократного исполнения.

**Итерация** — однократное выполнение тела цикла.

## 3o.1.1 Переменная, управляющая циклом

В заголовок любого цикла входит **переменная, управляющая циклом** (**параметр цикла**). Значение этой переменной определяет,
будет ли выполнено тело цикла.

Переменная, хранящая текущий номер итерации, называется **счётчиком итераций цикла** или просто **счётчиком цикла**.

«**счётчик цикла**» $\pmb{\not \equiv}$ «**параметр цикла**».

Счётчик цикла может отсутствовать, а параметр цикла всегда присутствует в цикле (и в заголовке, и в теле).

* * *

**Параметр цикла** проходит **три этапа**:

* 1. **Инициализация.**
Параметру цикла присваивается
начальное значение, прежде чем управление будет передано оператору
цикла. 

* 2. **Проверка.**
Значение параметра цикла проверяется перед каждой итерацией.

* 3. **Модификация.** 
На каждой итерации значение параметра цикла модифицируется.

### 3o.3.1. Безытерационный цикл

Тело цикла не выполняется, если условие повторения цикла при первой проверке оказалось ложным. Такой
цикл называют **безытерационным**.

**Безытерационный цикл не выполняется ни разу.**

**Пример 3o.1.**
В цикле увеличивать число $R$ на единицу при выполнении условия ${R\leqslant 0}$.

````python
# Возможен безытерационный цикл
R = int(input())
while R <= 0:
    R = R + 1
print(R)
````

Оператор цикла `R = R + 1` будет выполняться пока выполняется условие $R\leqslant 0$.

```
 Значение R | Цикл 
------------------------------------------- 
-5          | тело цикла будет выполняться; 
 5          | тело цикла ни разу не выполнится
```


### 3o.3.2. Вычисление конечных сумм и произведений


$$
\sum\limits_{i=1}^n a_i=a_1+a_2+{\dots}+a_n;\qquad
\prod\limits_{i=1}^n a_i=a_1\cdot a_2\cdot {\dots}\cdot a_n.
$$
$a_i$ зависит от постановки задачи.

**Алгоритм вычисления суммы**:

* 1. `s = 0`.
* 2. В цикле ($n$-раз для конечных сумм): `s = s + a`.


**Алгоритм вычисления произведения**:

* 1. `p = 1`.
* 2. В цикле ($n$-раз): `p = p * a`.

**Пример 3o.2.** Вычисление суммы первых 10 натуральных чисел с помощью цикла while:
$$
  1+2+3+{\dots}+10 = \sum\limits_{i=1}^{10} i.
$$

````python
sum_ = 0
i = 1           # инициализация
while i <= 10:  # проверка
    sum_ = sum_ + i
    i = i + 1   # модификация
print(sum_)
````

**Замечание.** Для решения задачи, конечно, не требуется цикл. Достаточно использовать формулу суммы $n$ первых  членов арифметической прогрессии:
$$
 S_n=\frac{a_1+a_n}{2}n.
$$

### 3o.3.3. Python как пакет символьных вычислений

**Пример 3o.3 (sympy).** Вычисление суммы первых 10 натуральных чисел с помощью методов пакета sympy:
$$
  1+2+3+{\dots}+10 = \sum\limits_{i=1}^{10} i.

````python
from sympy import *            # импорт всех функций из sympy

i = Symbol('i')                # i как символ

print(Sum(i, (i, 1, 10)))  # изображение суммы

s = Sum(i, (i, 1, 10)).doit()  # вычисление суммы

print(s)                       # на экране: 55
````

**Пример 3o.4 (sympy).** Вычисление суммы $\sum\limits_{i=1}^{10} \dfrac1i$ с помощью методов пакета sympy:

````python
from sympy import *

i = Symbol('i')

s1 = Sum(1/i, (i, 1, 10)).doit()          # 7381/2520
s2 = Sum(1/i, (i, 1, 10)).doit().evalf()  # 2.92896825396825

print(f'Результат - выражение: {s1}')
print(f'Результат - число    : {s2}')
````

### 3o.4.1 Числовые последовательности

**Последовательность** — это набор элементов некоторого множества, пронумерованный натуральными числами.

**Пример 3o.7** (из практикума).
Даны вещественные числа $a$ ($a>1$) и $b$. Получить и вывести на экран все члены последовательности $a^i$,
$i=1,2,3,\dots$, меньшие числа $b$.

**Решение.** Требуется найти степени числа $a$
$$
a, \  a^2, \ a^3,\dots
$$

Степени числа $a$ будем хранить в доп. переменной $d$.

````python
# a = 2 b = 10  2 4 8  d = a    d = d * a
````

````python
a = float(input('a (>1) >>> '))
b = float(input('b      >>> '))

if a < b:
    d = a
    while d < b:
        print('%5.2f' % d)
        d = d * a
else:
    print('Решения нет')
````

**Тест 1.**  Если $a\geqslant b$, то ни одного члена последовательности $a$, $a^2$, $a^3,\dots$, меньшего числа $b$ не существует. То есть условие цикла сразу ложно и тело цикла ни разу не выполнится.

**Тест 2.** Если $a<b$, то цикл начнет работу и, так как последовательность возрастающая, то условие продолжения цикла когда-нибудь нарушится и цикл закончит работу.

**Добавление операторов для печати трассировочной таблицы**

````python
yesTable = True  # True -- печать трассировочной таблицы

a = float(input('a (>1) >>> '))
b = float(input('b      >>> '))

if a < b:
    d = a
    if yesTable:  # операторы заголовка для трассы
        #print('%3s %8s' % ('d', 'd < b')) 
        #print('-' * 14) 
        print('%3d %8s' % (d, d < b))  # печать первой строки
    while d < b:
        if not yesTable:
            print('%5.2f' % d) 
        d = d * a
        if yesTable:
            print('%3d %8s' % (d, d < b))  # оператор печати для трассы
else:
    print('Решения нет')
````

**Пример 3o.8**. **Вариант A.** В последовательности целых чисел найти среднее арифметическое чисел, кратных 6.
Программа получает на вход количество чисел в последовательности, а затем сами числа.
В последовательности **всегда имеется число, кратное 6**. Программа должна вывести одно число — сумму чисел, кратных 6.


````python
# Параметр цикла i (изменяется от 1 до n)
n = int(input('n >>> '))

k = 0
s = 0
i = 1          # I
while i <= n:  # II
    a = int(input('a >>> '))
    if a % 6  == 0:
        k += 1
        s += a
    i += 1    # III
    
mean = s / k
print('Mean = %5.2f' % mean)
````

**Пример 3o.8. Вариант B.** В последовательности целых чисел найти среднее арифметическое чисел, кратных 6.  
Программа получает на вход количество чисел в последовательности, а затем сами числа.  
Программа должна вывести одно число — сумму чисел, кратных 6.  
***Если числа, кратные 6 отсутствуют, то вывести сообщение об этом.***

**Проблема:** сумма чисел, кратных 6, может быть равна 0.

````python
n = int(input('n >>> '))

k = 0
s = 0
i = 1
yes = False        # числа, кр.  6, ещё не встретились
while i <= n:
    a = int(input('a >>> '))
    if a % 6  == 0:
        k += 1
        s += a
        yes = True  # встретилось число, кр. 6
    i = i + 1

if yes: 
    mean = s / k
    print('Mean = %5.2f' % mean)
else: 
    print('Чисел нет')
````

**Пример3o.8. Вариант C.** В последовательности целых чисел найти среднее арифметическое чисел, кратных 6.  
**Количество чисел в последовательности неизвестно** (0 — признак окончания ввода чисел).  
В последовательности всегда имеется число, кратное 6.  
Программа должна вывести одно число — сумму чисел, кратных 6.

**Примечание.** Параметр цикла $a$ (пока $a \ne 0$).

````python
k = 0
s = 0

a = int(input('a >>> '))      # I
while a != 0:                 # II
    if a % 6 == 0:
        k += 1
        s += a
    a = int(input('a >>> '))  # IIII

mean = s / k

print('Mean = %5.2f' % mean)
print(f'Печать для контроля: sum = {s}, count = {k}')
````

**Пример 3o.9**. Дана последовательность $(a_i)_{i=1}^N$ $N$ целых неотрицательных чисел. Подсчитать количество элементов, следующих после первого нулевого.

Например, при $N=5$ возможны следующие наборы данных:
* 1) **0** 2 2 2 2 | 4 элемента
* 2) 1 3 **0** 4 5 | 2 элемента
* 3) 1 2 3 4 5     | 0 элементов 

**Условие окончания работы цикла**

`Условие1 = <достигнут конец послед-сти> ИЛИ <найден нулевой элемент>`

   
     
**Условие продолжения работы цикла  (Условие2 = not Условие1)**

`Условие2 = <НЕ достигнут конец послед.> И <НЕ найден 0-вой элемент>`

  
  
**Алгоритм**
```
while Условие2:   # что означает Условие2 == True:
     1) ввод элемента последовательности;
     2) наращивание счетчика элементов;
```

````python
n = 5

a = int(input('a >>> '))      # чтение I элемента
i = 1                         # номер I элемента
while i < n and a != 0:
    a = int(input('a >>> '))    # чтение текущего элемента
    i = i + 1                 # модификация счётчика

j = n - i
print(f'count = {j}')

# Тестовые примеры: 
'''
1) 0 2 2 2 2 | 4 элемента
2) 1 3 0 4 5 | 2 элемента
3) 1 2 3 4 5 | 0 элементов
'''

print()
````
