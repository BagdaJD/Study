**Решение с помощью вложенных условных операторов**

В каскадной условной инструкции условия if, …, elif проверяются по очереди.
При истинности какого-то из условий выполняется соответствующий блок. Если все
проверяемые условия ложны, то выполняется блок else, если он присутствует.

**Пример 3.14 из практикума**. Какой четверти принадлежит точка $(x, y)$?

````python
x = float(input('x >>> '))
y = float(input('y >>> '))

if x > 0 and y > 0:
    print('I четверть')
elif x < 0 and y > 0:
    print('II четверть')
elif x < 0 and y < 0:
    print('III четверть')
elif x > 0 and y < 0:
    print('IV четверть')
else:
    print(f'Точка ({x}, {y}) лежит на координатной оси', end=' ')
    print('или является началом координат.')
````

## 26s.2. Работа составных логических условий

```
<условие 1> or <условие 2>
<условие 1> and <условие 2>
```

**Сокращённая оценка** составных логических условий:

`<условие 2>` проверяется только, если от него зависит результат выражения, то есть, 

если `<условие 1>` **ложно** в выражении `<условие 1> or <условие 2>`;

если `<условие 1>` **истинно** в выражении `<условие 1> and <условие 2>`;


**Полная оценка** проверки составных логических условий: все условия проверяются независимо от результатов проверки предыдущих условий.

**Пример 26s.1.** Безаварийное вычисление возможно только при сокращенной оценке.

````python
x = int(input('x >>> '))

if x != 0 and 100 // x > 5:  
    print(100 // x)
else:
    print('x = 0 ИЛИ 100 // x <= 5')
````

Более **безопасный код** (не зависит от способе оценки логических выражений). Также **позволяет вывести более информативные сообщения об ошибке** (потребуется на контрольной работе).

````python
x = int(input('x >>> '))

if x != 0:
    if 100 // x > 5:  
        print(100 // x)
    else:
        print('100 // x <= 5')
else:
    print('x = 0')
````

**Пример 26.2.** Написать условный оператор для проверки того, что введённое натуральное число трёхзначное и его восьмеричная запись оканчивается на 3.

*Указание.* Для решения использовать функции `all()` и `any()`.

````python
# Будем сразу подставлять восьмеричные числа

s = '23'                   # строковое представление восьмеричного числ
x = int(s, 8)              # перевод восьмеричного числа в десятичное

print(f'x = {x} = {s}_8')  # для контроля входных данных

# ==> all()
if all([x >= 100, x <= 999, x % 8 == 3]):  # x >= 100 and x <= 999 and x % 8 == 3
    print('ok!')
else:
    print('not ok!')   
    
# ==> any
if any([x < 100, x > 999, x % 8 != 3]):  # x < 100 or x > 999 or x % 8 != 3
    print('not ok!')
else:
    print('ok!')  
````

**Пример 26s.4.**  Дано целое число $a$. Напечатать $\text{sign}\, a$.

$$
 \text{sign}\, a = \left\{
  \begin{array}{rl}
     1, & \text{если } x > 0; \\
     0, & \text{если } x = 0; \\
    -1, & \text{если } x < 0. 
  \end{array}
  \right.
$$

````python
x = int(input('x: int >>> '))

# Тернарный оператор
sgn1 = 1 if x > 0 else (0 if x == 0 else -1)   

   
#  Условный оператор (вложенные условные операторы)
if x > 0:
    sgn2 = 1
else:
    if x == 0:
        sgn2 = 0
    else:  
        sgn2 = -1

        
#  Каскадная условная конструкция
if x > 0:
    sgn3 = 1
elif x == 0:
    sgn3 = 0
else:  
    sgn3 = -1

print(f'sign({x}) = {sgn1}')
print(f'sign({x}) = {sgn2}')
print(f'sign({x}) = {sgn3}')
````

## 26s.3. Конструкция match/case (Python 3.10)

Конструкция match/case, которая называется **Structural pattern matching** (соответствие структуре шаблона).

Общий синтаксис конструкции `match/case`:
```
match <subject>:
    case <pattern_1>:
        <action_1>
    case <pattern_2>:
        <action_2>
    case <pattern_3>:
        <action_3>
    case _:
        <action_wildcard>
```
Оператор `match` принимает выражение `subject` и сравнивает его значение с последовательными шаблонами, заданными как один или несколько блоков `case`. Блок `case _` работает, как `else` (ни один из шаблонов не подходит).

Шаблоны в операторах case состоят из последовательностей, словарей, примитивных типов данных (int, float, str и т.д.), а также экземпляров классов.

**Пример 26s.5.**  Дан набор названий цветов светофора 
RED — красный
YELLOW — желтый
GREEN — зеленый
По первой буквы определить название цвета. 

````python
tlc = input('Traffic light color: str >>> ')

match tlc:
    case 'r' | 'R':
        res = 'Red'
    case 'y' | 'Y':
        res = 'Yellow'
    case 'g' | 'G':
        res = 'Green'
    case _:
        res = ':('

print(f'Traffic light color ({tlc}) = {res}')
````
