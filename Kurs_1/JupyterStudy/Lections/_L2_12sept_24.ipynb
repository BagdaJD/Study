# Введение в понятие функции

**Функция** — самостоятельный элемент программы, возвращающий некоторое значение.

-------------

Вызов любой функции имеет более высокий приоритет, чем все внешние относительно этого вызова
операции. Выражения, являющиеся аргументами вызываемой функции, вычисляются в момент вызова.

-------------

При вызове функции указывается имя функции и список фактических параметров (аргументов), заключенный в скобки:
```
y = math.sin(x)
z = math.exp(n * math.log(x))  # один из способов поиска x в степени n
print(divmod(17, 51))

```

**Кстати**

Способы нахождения $x^n$:

* с помощью операции `**`: `y = x**n` (Python, Perl);
* с помощью функции pow(): `p = pow(x, n)` (Python; C, C++);
* используя равенство $x^n = e^n \ln x$ ($x > 0$) (все ЯП, имеющие реализацию функции $\ln x$);
* с помощью циклов (все ЯП).



````python
# Пример злоупотребления ** и pow 
x = 5

print(x*x)
print(x**2)
print(pow(x, 2))
````

# Линейные алгоритмы

**Линейный алгоритм** — алгоритм, в котором отдельные предписания выполняются в порядке записи, независимо от значений исходных данных и промежуточных результатов.

**Пример Л.1 (вычисление по формулам. Пример из практикума).** 
Даны $x, y\in \mathbb{R}$. Вычислить
\begin{equation}\label{eq:1.2}
  u = \sqrt{ e^{x + y} },\quad
  v = -\frac{\ln( e^{x + y} + e - 1)}{2\sqrt{ e^{x + y} }} + x.
\end{equation}
Временная переменная ****t**** служит для однократного вычисления значения трижды повторяющегося выражения $t=e^{x + y}$.

````python
import math

x = float(input('x >>> '))
y = float(input('y >>> '))
print('x = %5.2f' % x)  # Эхо ввода (копирование текста из INPUT в OUTPUT)
print('y = %5.2f' % y)  # Эхо ввода

t = math.exp(x + y)  
u = math.sqrt(t)
v = -math.log(t + math.exp(1) - 1) / 2 / u + x

print('u = %5.2f' % u)  # Аннотированный вывод (размещение в OUTPUT поясняющего теста)
print('v = %5.2f' % v)  # Аннотированный вывод
````

# Проверка правильности программы

Любая программа должна пройти процесс отладки.

**Отладка** — работа по обнаружению и устранению ошибок программирования, а также доказательство правильности программы.

- Реальная задача 5%

## Продолжить ЗДЕСЬ!



Ошибки бывают *синтаксические* и *семантические* (смысловые)

**Семантические ошибки**
— ошибки, связанные с недостаточным знанием программистом языка программирования;
— логические ошибки и ошибки кодирования;
— ошибки при выполнении синтаксически правильных операторов (например, деление на нуль);
— ошибки, вызванные неверными данными.

Семантические ошибки могут проявиться на любом этапе решения задачи (от постановки задачи до сдачи готовой программы заказчику).

----

**Правильной** будем называть программу, которая успешно прошла этап трансляции (в ней нет синтаксических ошибок) и для всех
возможных входных данных обеспечивает получение правильного результата. 

----

Программа (алгоритм) **не является правильной**, если для некоторых исходных данных она дает неправильные результаты, сбои или не дает никаких результатов вообще.

----
Для проверки правильности программ используют **формальную верификацию** и **тестирование**.

**Формальная верификация** — доказательство с помощью формальных методов правильности/неправильности алгоритмов и программ в соответствии с формальным описанием их свойств.

**Тестирование** — проверка того, что программа решает именно поставленную задачу и выдает правильные результаты для любого набора возможных входных данных.

**Тестирование программы проводят**:

• в *нормальных условиях*: используют входные данные, типичные для эксплуатации программы в реальных условиях;

• в *экстремальных условиях*: используют входные данные, являющиеся «пограничными» для данной программы.

Для каждой программы экстремальные данные индивидуальны. Как правило, это очень малые или очень большие числа; числа, близкие

• в *исключительных условиях*: используют входные данные, значения которых лежат вне области определения функции; при выходе за

### Трассировка

Одним из способов выявления семантических ошибок является использование трассировочных таблиц.

**Трассировка** — это процесс пошаговой имитации на бумаге выполнения алгоритма компьютером.

Трассировочная таблица содержит указание значений переменных в виде таблицы, которые устанавливаются после выполнения
каждой команды присваивания или ввода.

**Обмен при помощи исключающего ИЛИ**

— алгоритм, который использует операцию XOR для обмена значений переменных, имеющих один и тот же тип данных без использования временной переменной.

В основе алгоритма лежит свойство: A XOR A = 0 для всех A.

Побитовые (поразрядные) логические операции выполняются отдельно над каждым разрядом (битом) двоичного числа независимо от разряда, стоящего справа или слева.

**Использование на практике**

Использование алгоритма распространено в ассемблерном коде для простых устройств (например, микроконтроллеров), в которых стоят
жёсткие требования к расходу памяти и тактов процессора.

На современных ЦП XOR-техника значительно медленнее, чем использование временной переменной для обмена. Это происходит по
причине распараллеливания выполнения команд. В XOR-технике операнды всех команд зависят от результатов предыдущих команд, поэтому они должны быть выполнены в строго последовательном порядке.

В каждом конкретном случае рекомендуется протестировать скорости обоих алгоритмов.
